```python
import matplotlib.pyplot as plt
import seaborn as sns

# Sort movies by weighted average
weight_average = movie_sorted_ranking.sort_values('weighted_average', ascending=False)

# Plot the top 10 movies based on weighted average
plt.figure(figsize=(12, 6))
axis1 = sns.barplot(
    x=weight_average['weighted_average'].head(10),
    y=weight_average['original_title'].head(10),
    data=weight_average,
    palette='husl'  # You can change the color palette here
)

plt.xlim(4, 10)
plt.title('Best Movies by Average Votes', weight='bold')
plt.xlabel('Weighted Average Score', weight='bold')
plt.ylabel('Movie Title', weight='bold')

# Save the plot as an image
plt.savefig('best_movies.png')
plt.show()
```

```python
popularity = movies_cleaned_df.sort_values('popularity', ascending=False)
popularity.head()
```

```python
plt.figure(figsize=(12, 6))
sns.barplot(
    x=popularity['popularity'].head(10),
    y=popularity['original_title'].head(10),
    data=popularity
)
plt.title('Most Popular by Votes', weight='bold')
plt.xlabel('Score of Popularity', weight='bold')
plt.ylabel('Movie Title', weight='bold')
plt.savefig('best_popular_movies.png')
```

```python
from sklearn.preprocessing import MinMaxScaler
scaling = MinMaxScaler()
movie_scaled_df = scaling.fit_transform(movies_cleaned_df[['weighted_average', 'popularity']])
movie_normalized_df = pd.DataFrame(movie_scaled_df, columns=['normalized_weight_average', 'normalized_popularity'])
```

```python
movies_cleaned_df[['normalized_weight_average', 'normalized_popularity']] = movie_normalized_df
```

```python
movies_cleaned_df['score'] = (
    movies_cleaned_df['normalized_weight_average'] * 0.5 +
    movies_cleaned_df['normalized_popularity'] * 0.5
)
```

```python

movies_scored_df = movies_cleaned_df.sort_values('score', ascending=False)
```

```python
print(movies_scored_df[['original_title', 'normalized_weight_average', 'normalized_popularity', 'score']].head(20))
```

```python
plt.figure(figsize=(16, 6))
sns.barplot(
    x=movies_scored_df['score'].head(10),
    y=movies_scored_df['original_title'].head(10),
    data=movies_scored_df,
    palette='deep'
)
plt.title('Best Rated & Most Popular Blend', weight='bold')
plt.xlabel('Score', weight='bold')
plt.ylabel('Movie Title', weight='bold')
plt.savefig('scored_movies.png')
```
